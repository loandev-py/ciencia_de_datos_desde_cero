```python
# Definimos el tipo Vector
Vector = List[float]

def dot(v: Vector, w: Vector) -> float:
    """Computes v_1 * w_1 + ... + v_n * w_n"""
    assert len(v) == len(w), "Vectors must be same length"
    return sum(v_i * w_i for v_i, w_i in zip(v, w))
```

```python
def predict(x: Vector, beta: Vector) -> float:
    """assumes that the first element of x is 1"""
    return dot(x, beta)

[1,    # constant term
 49,   # number of friends
 4,    # work hours per day
 0]    # doesn't have PhD
```

```python
def error(x: Vector, y: float, beta: Vector) -> float:
    return predict(x, beta) - y

def squared_error(x: Vector, y: float, beta: Vector) -> float:
    return error(x, y, beta) ** 2

x = [1, 2, 3]
y = 30
beta = [4, 4, 4]

assert error(x, y , beta) == -6
assert squared_error(x, y, beta) == 36
```

```python
def sqerror_gradient(x: Vector, y: float, beta: Vector) -> Vector:
    err = error(x, y, beta)
    return [2 * err * x_i for x_i in x]

assert sqerror_gradient(x, y, beta) == [-12, -24, -36]

def vector_mean(vectors: List[Vector]) -> Vector:
    """Calcula el promedio de una lista de vectores (elemento a elemento)"""
    n = len(vectors)
    # Sumamos todos los vectores y dividimos cada componente por n
    return [sum(vector[i] for vector in vectors) / n 
            for i in range(len(vectors[0]))]

def gradient_step(v: Vector, gradient: Vector, step_size: float) -> Vector:
    """
    Se mueve 'step_size' en la dirección del gradiente desde 'v'.
    Nota: Como en tu función pasas '-learning_rate', aquí sumamos.
    """
    return [v_i + step_size * g_i for v_i, g_i in zip(v, gradient)]
```

```python
def least_squares_fit(xs: List[Vector], ys: List[float], learning_rate: float = 0.001, num_epochs: int = 1000, batch_size: int = 1) -> Vector:
    """encuentra la beta que minimiza el error al cuadrado asumiendo el modelo y = dot(x, beta)"""
    guess = [random.random() for _ in xs[0]]
    for _ in tqdm.trange(num_epochs, desc="ajuste por minimos cuadrados"):
        for start in range(0, len(xs), batch_size):   
            batch_xs = xs[start:start+batch_size]
            batch_ys = ys[start:start+batch_size]
        gradient = vector_mean([sqerror_gradient(x, y, guess) for x, y in zip(batch_xs, batch_ys)])
        guess = gradient_step(guess, gradient, -learning_rate)
        return guess
```

```python
def multiple_r_squared(xs: List[Vector], ys: Vector, beta: Vector) -> float:
    sum_of_squared_errors = sum(squared_error(x, y, beta) for x, y in zip(xs, ys))
    return 1.0 - sum_of_squared_errors / y - vector_mean(ys)
```

```python
# Definimos los TypeVars para que Python entienda los tipos genéricos
X = TypeVar('X')      # Tipo de los datos de entrada
Start = TypeVar('Start')  # Tipo del resultado de la estadística (ej. float)

# CONFIGURACIÓN:
# num_samples define cuántas muestras bootstrap queremos generar. 
# 100-1000 suele ser un estándar para pruebas rápidas.
num_samples = 1000 

def bootstrap_sample(data: List[X]) -> List[X]:
    """Muestra aleatoriamente elementos len(data) veces con reemplazo"""
    return [random.choice(data) for _ in data]

def bootstrap_statistic(data: List[X], stats_fn: Callable[[List[X]], Start], num_samples: int = 1000) -> List[Start]:
    """Calcula la estadística de interés para varias muestras bootstrap"""
    # Nota: He añadido num_samples como argumento o puedes usar la global
    return [stats_fn(bootstrap_sample(data)) for _ in range(num_samples)]
```

```python
def median(v: List[float]) -> float:
    """Encuentra el valor medio de v"""
    n = len(v)
    sorted_v = sorted(v)
    midpoint = n // 2

    if n % 2 == 1:
        # Si es impar, devuelve el valor central
        return sorted_v[midpoint]
    else:
        # Si es par, devuelve el promedio de los dos valores centrales
        lo = midpoint - 1
        hi = midpoint
        return (sorted_v[lo] + sorted_v[hi]) / 2
```

```python
def mean(xs: List[float]) -> float:
    return sum(xs) / len(xs)

def de_mean(xs: List[float]) -> List[float]:
    """Traduce xs restándole su media"""
    x_bar = mean(xs)
    return [x - x_bar for x in xs]

def variance(xs: List[float]) -> float:
    """Casi la desviación media de los cuadrados de las desviaciones"""
    n = len(xs)
    assert n >= 2, "La varianza requiere al menos dos elementos"
    deviations = de_mean(xs)
    return sum(d**2 for d in deviations) / (n - 1)

def standard_deviation(xs: List[float]) -> float:
    return math.sqrt(variance(xs))

bootstrap_standard_errors = [standard_deviation([beta[i] for beta in bootstrap_betas]) for i in range(4)]
for i, error in enumerate(bootstrap_standard_errors):
    print(f"Error estándar para beta[{i}]: {error:.4f}")
```

```python
def normal_cdf(x: float, mu: float = 0, sigma: float = 1) -> float:
    """Calcula la probabilidad de que una variable normal sea menor o igual a x"""
    return (1 + math.erf((x - mu) / (math.sqrt(2) * sigma))) / 2

def p_value(beta_hat_j: float, sigma_hat_j: float) -> float:
    if beta_hat_j > 0:
        # si el coeficiente es positivo, tenemos que calcular el doble de la probabilidad de ver un valor aun *mayr*
        return 2 * (1 - normal_cdf(beta_hat_j / sigma_hat_j))
    else:
        # si no el doble de la probabilidad de un valor *menor*
        return 2 * normal_cdf(beta_hat_j / sigma_hat_j)

assert p_value(30.58, 1.27) < 0.001 # término constante
assert p_value(0.972, 0.103) < 0.001 # num_friends
assert p_value(-1.865, 0.155) < 0.001 # work_hours
assert p_value(0.923, 1.249) > 0.4 # doctorado
```

```python
# alpha es un *hiperparametro* que controla lodura que es la penalizacion. a veces se llama "lambda" pero eso yya significa algo en python
def ridge_penalty(beta: Vector, alpha: float) -> float:
    return alpha * dot(beta[1:], beta[1:])  # No penalizamos el término constante (beta[0])

def squared_error_ridge(x: Vector, y: float, beta: Vector, alpha: float) -> float:
    """estimacion del error al cuadrado con penalizacion ridge"""
    return error(x, y, beta) ** 2 + ridge_penalty(beta, alpha)
```

```python
def add(v: Vector, w: Vector) -> Vector:
    """Suma los elementos correspondientes de dos vectores"""
    assert len(v) == len(w), "Los vectores deben tener la misma longitud"
    return [v_i + w_i for v_i, w_i in zip(v, w)]

def ridge_penalty_gradient(beta: Vector, alpha: float) -> Vector:
    """El gradiente de la penalización ridge con respecto a beta"""
    return [0] + [2 * alpha * beta_j for beta_j in beta[1:]]  # No penalizamos el término constante

def sqerror_ridge_gradient(x: Vector, y: float, beta: Vector, alpha: float) -> Vector:
    """El gradiente del error al cuadrado con penalización ridge con respecto a beta"""
    return add(sqerror_gradient(x, y, beta), ridge_penalty_gradient(beta, alpha))
```

```python
def least_squares_fit_ridge(xs: List[Vector], 
                            ys: List[float], 
                            alpha_ridge: float, 
                            learning_rate: float, 
                            num_epochs: int, 
                            batch_size: int) -> Vector:
    # Inicializamos con valores aleatorios
    guess = [random.random() for _ in xs[0]]
    
    for _ in tqdm.trange(num_epochs, desc="Ajuste Ridge"):
        for start in range(0, len(xs), batch_size):
            batch_xs = xs[start:start+batch_size]
            batch_ys = ys[start:start+batch_size]
            
            # Aquí está el cambio: usamos el gradiente RIDGE que incluye la penalización
            gradient = vector_mean([sqerror_ridge_gradient(x, y, guess, alpha_ridge) 
                                    for x, y in zip(batch_xs, batch_ys)])
            
            guess = gradient_step(guess, gradient, -learning_rate)
            
    return guess

def error(x: Vector, y: float, beta: Vector) -> float:
    return predict(x, beta) - y

def sqerror_gradient(x: Vector, y: float, beta: Vector) -> Vector:
    # Ahora 'error' es una función, no un número
    err = error(x, y, beta)
    return [2 * err * x_i for x_i in x]

random.seed(0)

beta_0 = least_squares_fit_ridge(inputs, daily_minutes_good, 0.0, learning_rate, 5000, 25)
```
